Lets explore using mean, median and mode in python , with a fake income data revolving around 27,000 with a normal distribution and Standard Deviation of 15,000 and data points of 10,000 , computing the mean i.e average

```python
import numpy as np

incomes= np.random.normal(27000,15000,10000) # centered around, SD, data points
np.mean(incomes)
```

we can segment that data in 50 parts and plot it as a histogram

```python
import matplotlib.pyplot as plt
plt.hist(incomes, 50)
plt.show()
```

```python
# computing a median of the incomes will also bring the data close to 27k
np.median(incomes)
```

Now if we add a outlier value that can skew the data, we can observe the differences between mean and median , lets see it closer below

```python
incomes = np.append(incomes,[1000000000000])
```

```python
np.mean(incomes)
```

```python
np.median(incomes)
```

lets generate a random age data for around 500 people

```python
ages = np.random.randint(18, high= 91, size=500)
print(ages)
```

```python
ages = np.append(ages,[20])
```

```python
from scipy import stats
stats.mode(ages)
```
